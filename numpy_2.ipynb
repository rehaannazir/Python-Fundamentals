```python
#                                                         BroadCasting
import numpy as np
arr_1 = np.random.randint(1,51,size=(3,4))
arr_2 = np.random.randint(1,51,size=(3,1))
print(arr_1)
print(arr_2)
print(arr_1+arr_2)
```

```python
arr = np.random.randint(2,21,size=(3,4))
print('(3,4) \n',arr)
arr_1 = np.resize(arr,(6,2))
print('(6,2) \n',arr_1)
arr_2 = np.resize(arr,(12,1))
print('(12,1) \n',arr_2)
arr_3 = np.resize(arr,(8,2))  #Don't give any error
print('(8,2) \n',arr_3)
```

```python
#                                                         Transpose

arr = np.arange(12)
print(arr)
print(arr.transpose())
```

```python
arr = np.random.randint(2,12,size=(3,4))
print(arr)
print(arr.transpose())
```

```python
arr = np.random.randint(2,12,size=(3,1,2))
print(arr,arr.shape,end='\n',sep=' ')
new_arr = arr.transpose()
print(new_arr,new_arr.shape,end='\n',sep=' ')
```

```python
#                                             SwapAxes & Flatten
arr = np.random.randint(2,12,size=(3,1,2))
arr_2 = np.swapaxes(arr,0,2)
print(arr)
print(arr_2.shape,'\n',arr_2)
arr_3 = arr.flatten(order = 'C')
print(arr_3)
arr_3 = arr.flatten(order = 'F')
print(arr_3)
```

```python
#                                          Sorting
arr = np.random.randint(2,21,size = 10)
print(arr)
arr_2 = np.sort(arr,kind = 'quicksort')
print(arr_2)
arr_3 = np.sort(arr,kind = 'headsort')
print(arr_3)
arr_4 = np.sort(arr,kind = 'mergeksort')
print(arr_4)
```

```python
arr = np.random.randint(2,21,size =(4,3))
print(arr)
arr_2 = np.sort(arr,axis=None)
print(arr_2)
arr_3 = np.sort(arr,axis=0)
print(arr_3)
arr_4 = np.sort(arr,axis=1)
print(arr_4)
```
